```python
# Handling outliers
# we can use two methods to handle the outliers
# 1 = z-score  , 2 = IQR method
```

```python
# Z-Score method
from scipy import stats
z_score_age = np.abs(stats.zscore(df['Income']))
z_score_income = np.abs(stats.zscore(df['Income']))

# indetifieng them
outliers_age = df[z_score_age > 3]
outliers_income = df[z_score_income > 3]

print("Outliers in Age column:")
print(outliers_age)
print('*************')
print("\nOutliers in Income column:")
print(outliers_income)
```

```python

# with IQR method
Q1_age = df['Age'].quantile(0.25)
Q3_age = df['Age'].quantile(0.75)
IQR_age = Q3_age - Q1_age

# with income
Q1_income = df['Income'].quantile(0.25)
Q3_income = df['Income'].quantile(0.75)
IQR_income = Q3_income - Q1_income

# define the outliers treshold
lower_bound_age = Q1_age - 1.5 * IQR_age
upper_bound_age = Q3_age + 1.5 * IQR_age

lower_bound_income = Q1_income - 1.5 * IQR_income
upper_bound_income = Q3_income + 1.5 * IQR_age

# identify the outliers
outliers_age_by_iqr= df[(df['Age'] < lower_bound_age) | (df['Age'] > upper_bound_age)]
outliers_income_by_iqr = df[(df['Income'] < lower_bound_income) | (df['Income'] > upper_bound_income)]

print("Outliers in Age column:")
print(outliers_age_by_iqr)
print('*************')
print("\nOutliers in Income column:")
print(outliers_income_by_iqr)
```

```python
# summarizing the result of statics
stats_summary = pd.DataFrame({
    'Variable': ['Year_Birth', 'Income'],
    'Mean': [age_mean, income_mean],
    'Median': [age_median, income_median],
    'Mode': [age_mode, income_mode],
    'Std Dev': [age_std, income_std],
    'Variance': [age_var, income_var],
    'Range': [age_range, income_range]
})

stats_summary
```

```python
# Handling outliers
# removing and adjusting the outliers
df_cleaned = df[~df.index.isin(outliers_age_by_iqr.index)]
df_cleaned = df_cleaned[~df_cleaned.index.isin(outliers_income_by_iqr.index)]
```

## Task 3 - Probability Distributions üé≤

- **Identify variables that could follow specific probability distributions (e.g., Binomial, Normal). Calculate probabilities and expected values based on these distributions.**



```python
# Calculating Probability and Expected Values
import scipy.stats as stats

# parameters for the binomial distribution
n_trials = df.shape[0]    # total customers
p_success = df['Response'].mean()   # probability of accepting the offer

# expected success
expected_success = n_trials * p_success
print(f"Expected number of successes for 'Response': {expected_success}")

# lets suppose we are calculating PMF for 100 trials
k = 100
probability_of_k_success = stats.binom.pmf(k, n_trials, p_success)
print(f"Probability of getting {k} successes: {probability_of_k_success}")
```

```python
# for cumulative prbability
cumulative_probability = stats.binom.cdf(k, n_trials, p_success)
print(f"Cumulative probability of getting {k} or fewer successes: {cumulative_probability}")
```

```python
# Normal Distribution
# calculating parameters
# for income
print(f"Mean income: {income_mean}")
print(f"Standard deviation of income: {income_std}")

# for Age
print(f"Mean age: {age_mean}")
print(f"Standard deviation of age: {age_std}")
```

```python
# calculating the probabilities
# let suppose the threshold value of the income is 50000 and we want to get the probabily of income less then the threshold value
threshold_income = 50000
probability_less_then_threshold = stats.norm.cdf(threshold_income , income_mean,income_std)
print(f"Probability of income being less than {threshold_income}: {probability_less_then_threshold}")

# probability of age is between 30 to 50
lower_age_threshold = 30
upper_age_threshold = 50
probability_age_between_thresholds = stats.norm.cdf(upper_age_threshold, age_mean, age_std) - stats.norm.cdf(lower_age_threshold, age_mean, age_std)
print(f"Probability of age being between {lower_age_threshold} and {upper_age_threshold}: {probability_age_between_thresholds}")
```

## Task 4: Insights and Customer Segmentation üìà

- **Explore relationships between customer characteristics and spending habits. Segment customers based on their behaviors and characteristics.**

```python
# Lets caculate the correlation for get the relationship between the variables
numeric_columns = df[['Income', 'MntWines', 'MntFruits', 'MntMeatProducts','MntFishProducts', 'MntSweetProducts', 'MntGoldProds']]
correlation_matrix =numeric_columns.corr()
correlation_matrix
```

```python
# Spending Pattern Demographics
# lets use the group by for better understanding that how the spending is diffrent by the customers demographics (Age,Education,Marital_Stats)
spending_by_age = df.groupby('Age')[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].mean()
spending_by_education = df.groupby('Education')[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].mean()
spending_by_marital_status = df.groupby('Marital_Status')[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].mean()

print("Spending by Age:")
print(spending_by_age)
print("\nSpending by Education:")
print(spending_by_education)
print("\nSpending by Marital Status:")
print(spending_by_marital_status)
```

```python
# We can use visualization for better understanding
# spending by education
spending_by_education.plot(kind='bar')
plt.title('Spending by Education')
plt.xlabel('Education')
plt.ylabel('Average Spending')
plt.show()

# spending by marital status
spending_by_marital_status.plot(kind='bar')
plt.title('Spending by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Average Spending')
plt.show()

# spending by age
spending_by_age.plot(kind='bar', figsize=(12, 8))
plt.title('Average Spending by Age')
plt.xlabel('Age')
plt.ylabel('Average Spending')
plt.show()
```

```python
# Customer Segmentation
# we can use the clutering technique for the segmentation
# K-Means
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# dataset for clustering
variables = df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']]

# standardize the scaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(variables)     # scaling for the data to get a better result

# Applying the K-Means Clustering
kmeans = KMeans(n_clusters = 4,random_state = 42)   # total 4 clusters we are making here
df['Cluster'] = kmeans.fit_predict(scaled_data)

# Clusters Center
clusters_centroids = scaler.inverse_transform(kmeans.cluster_centers_)      # inverse_transform will convert back to the starting scale of the data and give the exact result
print( "Clusters Centers ", clusters_centroids)
```

```python
# lets visualize the clustered data
import matplotlib.pyplot as plt
import seaborn as sns

# for income and wines
plt.figure(figsize = (6,6))
sns.scatterplot(data = df, x = 'Income',y = 'MntWines',hue = 'Cluster',palette= 'viridis')
plt.title('Customers Segmentation On The Bases Of Income And Wine Spending')
plt.xlabel('Income')
plt.ylabel('MntWines')
plt.show()

# for income and fruits
plt.figure(figsize = (6,6))
sns.scatterplot(data = df, x = 'Income',y = 'MntFruits',hue = 'Cluster',palette= 'viridis')
plt.title('Customers Segmentation On The Bases Of Income And Fruits Spending')
plt.xlabel('Income')
plt.ylabel('MntFruits')
plt.show()

# for income and meat products
plt.figure(figsize = (6,6))
sns.scatterplot(data = df, x = 'Income',y = 'MntMeatProducts',hue = 'Cluster',palette= 'viridis')
plt.title('Customers Segmentation On The Bases Of Income And Meat Products Spending')
plt.xlabel('Income')
plt.ylabel('MntMeatProducts')
plt.show()

# for income and fish products
plt.figure(figsize = (6,6))
sns.scatterplot(data = df, x = 'Income',y = 'MntFishProducts',hue = 'Cluster',palette= 'viridis')
plt.title('Customers Segmentation On The Bases Of Income And Fish Spending')
plt.xlabel('Income')
plt.ylabel('MntFishProducts')
plt.show()

# for income and sweets products
plt.figure(figsize = (6,6))
sns.scatterplot(data = df, x = 'Income',y = 'MntSweetProducts',hue = 'Cluster',palette= 'viridis')
plt.title('Customers Segmentation On The Bases Of Income And Sweets Spending')
plt.xlabel('Income')
plt.ylabel('MntSweetProducts')
plt.show()

# for income and gold products
plt.figure(figsize = (6,6))
sns.scatterplot(data = df, x = 'Income',y = 'MntGoldProds',hue = 'Cluster',palette= 'viridis')
plt.title('Customers Segmentation On The Bases Of Income And Gold Products Spending')
plt.xlabel('Income')
plt.ylabel('MntGoldProds')
plt.show()
```
